```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
%matplotlib inline
```

```python
cylinders = data['cylinders']
a = data.groupby('cylinders').size()

y_axis = np.array(a)
x_axis = np.array(a.index[:])

ind = np.arange(len(y_axis))

plt.bar(ind,y_axis)
plt.xticks(ind,x_axis) 
plt.title('Cylinder Distribution')
```

# Q.8 -Create a bar chart that shows the distribution for model year?
